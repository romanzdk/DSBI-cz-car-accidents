# 00 Data joining

```python
import pandas as pd
import glob
import warnings
warnings.filterwarnings('ignore')
```

```python
# import headers
headers = pd.read_excel('hlavicky.xlsx').iloc[30]
```

```python
# select all files matching criteria
all_filenames = [i for i in glob.glob('01_2019_*.csv')]
```

```python
#combine all files in the list + add header
combined_csv = pd.concat([pd.read_csv(f, encoding='cp1250', sep=';', low_memory=False, header=None, names=headers) for f in all_filenames], sort=False)
```
